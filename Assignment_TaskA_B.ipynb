```python
from pymongo import MongoClient
from pprint import pprint
import csv

# Create a MongoClient
client = MongoClient()

# We will use the database: fit5148
db = client.fit5148_db

# The collection name is units
hsp = db.hotspots_his
clm = db.climate_his

if __name__ == "__main__":
   print("This is main function")
```

```python
db.hotspots_his.drop()
csvfile = open('hotspot_historic.csv', 'r')
reader = csv.DictReader(csvfile)
header = ["latitude","longitude","datetime","confidence","date","surface_temperature_celcius"]
for each in reader:
    row={}
    for field in header:
        if field=='latitude':
            each[field] = float(each[field])
        if field=='longitude':
            each[field] = float(each[field])
        if field=='confidence':
            each[field] = int(each[field])
        if field=='surface_temperature_celcius':
            each[field] = int(each[field])
        row[field]=each[field]
    db.hotspots_his.insert(row)
```

```python
db.climate_his.drop()
csvfile = open('climate_historic.csv', 'r')
reader = csv.DictReader(csvfile)
header = ["station","date","air_temperature_celcius","relative_humidity","windspeed_knots","max_wind_speed","precipitation "]
for each in reader:
    row={}
    for field in header:
        if field=='air_temperature_celcius':
            each[field] = int(each[field])
        if field=='station':
            each[field] = int(each[field])
        if field=='max_wind_speed':
            each[field] = float(each[field])
        if field=='relative_humidity':
            each[field] = float(each[field])
        if field=='windspeed_knots':
            each[field] = float(each[field])
        row[field]=each[field]
    db.climate_his.insert(row)
```

```python
db.clim_hot_model.drop()
model = db.clim_hot_model
for each in clm.find():
    hot_spot_records = []
    results = hsp.find({"date":each["date"]})
    for data in results:
        hot_spot_records.append(data)
    each["hot_spot_records"] = hot_spot_records
    db.clim_hot_model.insert(each)
```

### 2.

#### a. Find climate data on 10th of Dec 2017

```python
for i in clm.find({'date':'10/12/2017'}):
    pprint(i)
```

#### b. Find the latitude, longitude, surface temperature and confidence when the surface temperature was between 65 and 100

```python
for i in hsp.find({"$and":[{'surface_temperature_celcius':{"$lt":100}},{'surface_temperature_celcius':{"$gt":65}}]},
                                    {'latitude':1,'longitude':1,'surface_temperature_celcius':1,'confidence':1}):
    pprint(i)
```

#### c. Find date, surface temperature, air temperature, relative humidity and max wind speed on 15th and 16th of Dec, 2017

```python
for i in db.clim_hot_model.find({"$or":[{'date':'15/12/2017'},{'date':'16/12/2017'}]},
                                    {'date':1,'air_temperature_celcius':1,'relative_humidity':1,'max_wind_speed':1,'_id':1,'hot_spot_records.surface_temperature_celcius':1}):
    pprint(i)
```

#### d. Find datetime, air temperature, surface temperature and confidence when confidence is between 80 and 100

```python
for i in hsp.find({"$and":[{'confidence':{"$lt":100}},{'confidence':{"$gt":80}}]},
                                    {'date':1,'surface_temperature_celcius':1, 'confidence':1,'_id':1}):
    hot_date = str(i['date'])
    del i['date']
    pprint(i)
    for j in clm.find({'date':{"$eq":hot_date}},{'air_temperature_celcius':1}):
        del j['_id']
        pprint(j)
        
```

#### e. Find the top 10 records with the highest surface temperature

```python
for i in hsp.aggregate([{"$sort" : {'surface_temperature_celcius':-1}},{ "$limit" : 10 }]):
    pprint(i)
```

#### f. Find the number of fire each day. You are required to only display the total number of fire and the date in the output.

```python
result = hsp.aggregate([{"$group":{"_id" :"$date" , "count": {"$sum":1}}}])
for document in result:
    pprint(document)
```

#### g. Find the average surface temperature for each day. You are required to only display average surface temperature and the date in the output.

```python
results = hsp.aggregate([{"$group":{"_id":"$date", "average_surface_temperature_celcius":{"$avg":"$surface_temperature_celcius"}}}])
for document in results:
    pprint(document)
```
